```python
import pandas as pd
from collections import Counter
```

```python
hole_df = pd.read_excel("BEBA_U2_EA_BERC.xlsx")
profundidad = hole_df['depth']
```

```python
data = profundidad.drop(0)

print(f'Promedio: {data.mean()}')
print(f'Mediana: {data.median()}')
print(f'Moda: {data.mode()[0]}')
print()
print(f'Desviacion: {data.std(ddof=1)}')
print(f'Varianza: {data.var(ddof=1)}')
print()
print(profundidad.size)
```
